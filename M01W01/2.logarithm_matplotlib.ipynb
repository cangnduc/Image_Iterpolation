#### Gaussian

```python
import math
import matplotlib.pyplot as plt

def gaussian(x, mean, variance):
    part1 = 1 / math.sqrt(2*math.pi*variance)
    part2 = math.exp( -(1/(2*variance)) * (x-mean)**2 )
    return part1*part2
```

```python
mean = 10
variance = 4

data = list(range(-10, 30+1))
f_data = [gaussian(x, mean, variance) for x in data]

# plot
plt.plot(data, f_data)
plt.show()
```

```python
mean = 10
variance = 4

k = 5
data = list(range(-10*k, 30*k+1))
data = [x/k for x in data]

f_data = [gaussian(x, mean, variance) for x in data]

# plot
plt.plot(data, f_data)
plt.show()
```

#### Log_gaussian

```python
import math
import matplotlib.pyplot as plt

def gaussian(x, mean, variance):
    part1 = 1 / math.sqrt(2*math.pi*variance)
    part2 = math.exp( -(1/(2*variance)) * (x-mean)**2 )
    return part1*part2
```

```python
mean = 10
variance = 4

k = 5
data = list(range(-10*k, 30*k+1))
data = [x/k for x in data]

f_data = [gaussian(x, mean, variance) for x in data]
f_data = [math.log(x) for x in f_data]

# plot
plt.plot(data, f_data)
plt.show()
```

#### Parabolic Function and its log function

```python
import math
import matplotlib.pyplot as plt

k = 10
data = list(range(-5*k, 5*k+1))
data = [x/k for x in data]
```

```python
squared_data = [x**2 + 1e-3 for x in data]
plt.plot(data, squared_data)
plt.show()
```

```python
log_squared_data = [math.log(x) for x in squared_data]
plt.plot(data, log_squared_data)
plt.show()
```
